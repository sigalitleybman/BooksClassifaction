```python
load_books_df = pd.read_csv("final_books_dataframe.csv", index_col = [0])
```

```python
load_books_df.info()
```

```python
""" made a copy of our original dataframe"""

books_df_copy = load_books_df.copy()
```

```python
""" dropped rows that have at least one NaN """

books_df_copy.dropna(axis = 0, inplace = True)
```

```python
""" dropped duplicated rows """

books_df_copy.drop_duplicates(inplace = True)
```

```python
books_df_copy.info()
```

```python
# We just realized that pandas decided to import some of our needed columns as a string (object)
# so we're going to convert them back to numeric
# the columns are: book_length_mm, book_width_mm, book_weight_grams, bestsellers_rank

incorrect_columns_types = ['book_length_mm', 'book_width_mm', 'book_weight_grams', 'bestsellers_rank']
```

```python
# getting rid of commas in these object columns, and converting the type to the accurate one.
    
books_df_copy['bestsellers_rank'] = books_df_copy['bestsellers_rank'].str.replace(',', '')
books_df_copy['bestsellers_rank'] = books_df_copy['bestsellers_rank'].astype(int)
books_df_copy.info()    
```

```python
# getting rid of commas in these object columns, and converting the type to the accurate one.

books_df_copy['book_weight_grams'] = books_df_copy['book_weight_grams'].str.replace(',', '')
books_df_copy['book_weight_grams'] = books_df_copy['book_weight_grams'].astype(float)
books_df_copy.info()    
```

```python
# getting rid of commas in these object columns, and converting the type to the accurate one 

books_df_copy['book_length_mm'] = books_df_copy['book_length_mm'].str.replace(',', '')
books_df_copy['book_length_mm'] = books_df_copy['book_length_mm'].astype(float)


books_df_copy['book_width_mm'] = books_df_copy['book_width_mm'].str.replace(',', '')
books_df_copy['book_width_mm'] = books_df_copy['book_width_mm'].astype(float)
```

```python
# replace str columns to catgorical

replace_month_col_to_catgorical = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8,
                            'Sep':9,'Oct':10, 'Nov':11, 'Dec':12}

replace_genre_col_to_catgorical = {'contemporary fiction' : 1, 'sport' : 2 , 'mind,body & spirit': 3 }

replace_language_col_to_categorical = {'English':1, 'Spanish':2 , 'French':3 , 'Italian':4, 'German':5, 'Latin':6}

books_df_copy['publication_month'].replace(replace_month_col_to_catgorical, inplace=True)
books_df_copy['book_genre'].replace(replace_genre_col_to_catgorical, inplace=True)
books_df_copy['language'].replace(replace_language_col_to_categorical, inplace=True)

books_df_copy.head()
```
