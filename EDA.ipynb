## Let's explore the pairplot of the 

```python
sns.pairplot(df_copy[features])
plt.grid()
```

As we can see, there are some moderate positive correlations between different columns,

book_length_mm and book_width_mm,

#### we wanted to plot all the publishing countries that appear in our dataframe, and to mark all the coutrnies we've visited ourselves in one color and other countries in another

```python
# first we had to get a list of longitudes and latitudes for each of the appearing countries 

countries = []
longitude = []
latitude = []
```

```python
#saving resources

session = requests.Session()
```

**we registered to https://developer.mapquest.com/ and asked for api key inorder to use their api,<br>
and obtain the longitude and latitude for each of the countries**

```python
url = "http://www.mapquestapi.com/geocoding/v1/address"
key = "R5WyVOezDdHMq448pibt4AxqGyBoKAFF"
```

```python
""" we had to clean the data a little bit"""

ans = df_copy['publication_country'].str.replace(']','')
ans = ans.str.replace('[','')
ans = ans.str.replace('.', '')
```

**extracted the latitude and longitude for each of the appeared countries in our data**

```python
for country in ans.unique():
    
    parameters = {
        "key" : key,
        "location" : country
    }
    
    response = session.get(url, params= parameters) 
    
    data = json.loads(response.text)['results'][0]['locations'][0]['latLng']
   
    countries.append(country)
    latitude.append(data['lat'])
    longitude.append(data['lng'])
```

**created new dataframe "df_latLng"**

```python

latLng_data = {'country': countries, 'latitude': latitude, 'longitude' : longitude }
df_latLng = pd.DataFrame(data=latLng_data)
df_latLng
```

**we wanted to color the countries we've visited, so we had to add a new col to the new df which indicates the countries we've visited**

```python
visited_countries_by_us = ['Canada', 'Spain', 'France', 'Italy','Germany', 'IL', 'VA', 'S.I.']
```

**the function sets the right color to each of the values**

```python
def did_visit(row):
    if row['country'] in visited_countries_by_us:
        return 'firebrick'
    return 'darkorchid'
```

```python
df_latLng['color'] = df_latLng.apply(did_visit, axis=1)
df_latLng.head()
```
