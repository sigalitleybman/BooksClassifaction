# Outliers

### Remove outliers according to IQR

```python
outliers_columns = ['book_rate', 'number_of_pages', 'book_length_mm', 'book_width_mm', 'book_depth_mm',
                    'book_weight_grams', 'publication_year', 'bestsellers_rank' ,'price_nis']
```

```python
# plot distributions

sns.set_style('ticks',{'axes.grid' : False})

for f in outliers_columns:
    fig, (ax) = plt.subplots(1, 1, figsize=(10,7))
    
    
    ax.boxplot(df_copy[f], vert=False)
    ax.grid()   
    ax.set_title('Boxplot of ' + f)
    
    plt.show()
```

```python
outlier_indexes_list = []
for col in outliers_columns:
    Q1 = np.percentile(df_copy[col], 25)
    Q3 = np.percentile(df_copy[col], 75)
    IQR = Q3 - Q1
    outlier_indexes = df_copy[(df_copy[col] < Q1 - 1.5*IQR) | (df_copy[col] > Q3 + 1.5*IQR)].index
    outlier_indexes_list.extend(outlier_indexes)
    print("column is:" , col, "\nlen is: " ,len(outlier_indexes))
    print('\n')
    
# Because it is possible to have duplicate rows, we will create a set    
index_set = set(outlier_indexes_list)
print(len(index_set))
```

```python
df_copy.shape
```

```python
# Remove the outliers indexes and their rows

df_copy.drop(index_set, inplace=True)
```

```python
df_copy.shape
```

```python
# as seen above at the description of df_copy , there are still outliers like, publication year > 2022.
# so it must be dealt by omitting the corresponding relevant rows.
# we took into consideration the books that were published up to feb 2022. 


df_year_greater_than_2022 = df_copy[df_copy["publication_year"] >= 2022]
df_month_greater_than_1 = df_year_greater_than_2022[df_year_greater_than_2022["publication_month"] > 1]
```

```python
# made another copy of the copied df
df_copy_2 = df_copy.copy()
```

```python
# dropped the outliers

df_copy_2.drop(df_month_greater_than_1.index, inplace = True) 
```
